## Feature Engineering

```python
df.head()
```

```python
df.drop(["Unnamed: 0"] , axis = 1 , inplace = True)
```

```python
df["Gender"].value_counts()
```

```python
gender_labels = {
    "Male" : 1,
    "Female" : 0,
}
df['Gender_labels'] = df['Gender'].map(gender_labels)
df.drop("Gender" , axis = 1 , inplace = True)
```

```python
df["Location"].value_counts()
```

```python
locations = {
    "Suburban" : 0,
    "Urban" : 1,
    "Rural" : 2
}

df['Location_labels'] = df['Location'].map(locations)
df.drop("Location" , axis = 1 , inplace = True)
```

```python
df["Interests"].value_counts()
```

```python
interests = {
    "Sports": 0,
    "Fashion": 1,
    "Travel": 2,
    "Food": 3,
    "Technology": 4
}
df['Interests_labels'] = df['Interests'].map(interests)
df.drop("Interests" , axis = 1 , inplace = True)
```

```python
df["Product_Category_Preference"].value_counts()
```

```python
product_category_labels = {
    "Apparel": 0,
    "Electronics": 1,
    "Books": 2,
    "Home & Kitchen": 3,
    "Health & Beauty": 4
}

df["Product_Category_Preference"] = df["Product_Category_Preference"].map(product_category_labels)
df = df.replace({True: 1, False: 0})
```

```python
df.head()
```

```python
# Min Max Scaler on Income and Average_Order_Value

import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import joblib

# Assuming df is your DataFrame containing the data

# Train MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(df[['Income', 'Total_Spending']])

# Transform training data
df[['Income', 'Total_Spending']] = scaler.transform(df[['Income', 'Total_Spending']])

# Save scaler model
joblib.dump(scaler, "min_max_scaler.pkl")
```

```python
df_train = df
df_train["User_ID"] = df_train['User_ID'].str.replace('#', '').astype(int)
```

## Create 100 clusters

#### UserIDs lying in the same cluster will be shared the products which the people belonging to the same cluster bought

```python
from sklearn.cluster import KMeans
import numpy as np

# Initialize KMeans model
kmeans = KMeans(n_clusters=100, random_state=42)

# Fit the model
kmeans.fit(df_train)

# Get cluster labels for each user ID
cluster_labels = kmeans.predict(df_train)

# Create a dictionary to store user ID and corresponding cluster ID
user_cluster_dict = dict(zip(df["User_ID"], cluster_labels))
```
