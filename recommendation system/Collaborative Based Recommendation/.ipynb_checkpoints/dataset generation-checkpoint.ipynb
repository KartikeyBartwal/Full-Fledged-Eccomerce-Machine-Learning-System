```python
import pandas as pd
import numpy as np

# Generate synthetic data
np.random.seed(42)

# Number of samples
num_samples = 1000

# Features representing user personalization and activities
data = {
    'User_ID': ['#' + str(i) for i in range(1, num_samples + 1)],
    'Age': np.random.randint(18, 65, num_samples),  # User's age
    'Gender': np.random.choice(['Male', 'Female'], num_samples),  # User's gender
    'Location': np.random.choice(['Urban', 'Suburban', 'Rural'], num_samples),  # User's location
    'Income': np.random.randint(20000, 150000, num_samples),  # User's income
    'Interests': np.random.choice(['Fashion', 'Technology', 'Sports', 'Travel', 'Food'], num_samples),  # User's interests
    'Last_Login_Days_Ago': np.random.randint(1, 30, num_samples),  # Days ago since last login
    'Purchase_Frequency': np.random.randint(0, 10, num_samples),  # Number of purchases in the last month
    'Average_Order_Value': np.random.randint(10, 200, num_samples),  # Average order value in dollars
    'Total_Spending': np.random.randint(100, 5000, num_samples),  # Total spending in the last month
    'Product_Category_Preference': np.random.choice(['Electronics', 'Apparel', 'Home & Kitchen', 'Books', 'Health & Beauty'], num_samples),  # Preferred product category
    'Time_Spent_on_Site_Minutes': np.random.randint(1, 600, num_samples),  # Time spent on the site in minutes
    'Pages_Viewed': np.random.randint(1, 50, num_samples),  # Number of pages viewed in the last session
    'Newsletter_Subscription': np.random.choice([True, False], num_samples),  # Subscription to newsletter
}

# Create DataFrame
df = pd.DataFrame(data)

# Display the DataFrame
print(df.head())
```
